```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 안 깨지게 하는 코드
from matplotlib import rc
rc('font', family='Malgun Gothic') # Windows
plt.rcParams['axes.unicode_minus'] = False

#데이터셋
excel_file_path = 'C:\\Users\\rkffl\\SubwayPulse\\1.csv'

# ---------- 여기! 파일을 읽어오는 부분을 이렇게 다시 살려줘! ----------
try:
    df = pd.read_csv(excel_file_path, encoding='cp949')
    print("CSV 파일을 cp949 인코딩으로 성공적으로 읽었습니다.")
except UnicodeDecodeError:
    print("cp949 인코딩으로 파일을 읽는 데 실패했습니다. utf-8로 다시 시도합니다.")
    df = pd.read_csv(excel_file_path, encoding='utf-8')
    print("CSV 파일을 utf-8 인코딩으로 성공적으로 읽었습니다.")
# ---------------------------------------------------------------------

# pandas 설정 변경: 모든 행과 모든 열을 출력하도록 설정
pd.set_option('display.max_rows', None)     # 최대 행 수 제한 해제 (모든 행 출력)
pd.set_option('display.max_columns', None)  # 최대 열 수 제한 해제 (모든 열 출력)
pd.set_option('display.width', 1000)        # 출력 너비 설정


print("\n전체 데이터프레임 내용 (상위 5줄):")
print(df.head()) # 전체 다 출력하면 너무 기니까, head만 출력할게


# '호선명'을 기준으로 그룹화하고, 각 그룹의 '승차총승객수' 합계 구하기
df_grouped_sum = df.groupby('호선명')['승차총승객수'].sum().reset_index()
df_grouped_sum2 = df.groupby('호선명')['하차총승객수'].sum().reset_index()

# 호선별 승/하차총승객수 합계를 하나의 DataFrame으로 합치기
df_total_sum = pd.merge(df_grouped_sum, df_grouped_sum2, on='호선명')

# '호선명'을 기준으로 그룹화하고, 여러 컬럼의 평균 구하기
df_grouped_avg = df.groupby('호선명')[['승차총승객수', '하차총승객수']].mean().reset_index()


#승차합
print("\n--- 호선별 승차총승객수 합계 ---")
print(df_grouped_sum.sort_values(by='승차총승객수', ascending=False))

#하차합
print("\n--- 호선별 하차총승객수 합계 ---")
print(df_grouped_sum2.sort_values(by='하차총승객수', ascending=False))


# '호선명'을 기준으로 그룹화하고, 여러 컬럼의 평균 구하기
print("\n--- 호선별 승하차총승객수 평균 ---")
print(df_grouped_avg.sort_values(by='승차총승객수', ascending=False))
    
```

```python
#데이터 가공

# 1. 호선별 승차총승객수 합계 막대 그래프
plt.figure(figsize=(15, 7)) # 그래프 크기 설정
sns.barplot(x='호선명', y='승차총승객수', data=df_grouped_sum.sort_values(by='승차총승객수', ascending=False))
plt.title('호선별 승차총승객수 합계', fontsize=18)
plt.xlabel('호선명', fontsize=12)
plt.ylabel('승차총승객수', fontsize=12)
plt.xticks(rotation=45, ha='right') # 호선명 라벨이 겹치지 않게 회전
plt.grid(axis='y', linestyle='--', alpha=0.7) # 그리드 라인 추가
plt.tight_layout() # 그래프 요소들이 잘 보이도록 자동 조정
plt.show() # 그래프 출력

    # 2. 호선별 하차총승객수 합계 막대 그래프
plt.figure(figsize=(15, 7))
sns.barplot(x='호선명', y='하차총승객수', data=df_grouped_sum2.sort_values(by='하차총승객수', ascending=False))
plt.title('호선별 하차총승객수 합계', fontsize=18)
plt.xlabel('호선명', fontsize=12)
plt.ylabel('하차총승객수', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 3. 호선별 승/하차총승객수 합계 비교 막대 그래프 (두 개의 막대)
# 데이터를 long format으로 변환하여 seaborn이 여러 그룹을 한 번에 그릴 수 있도록 함
df_total_sum_melted = df_total_sum.melt(id_vars='호선명', var_name='구분', value_vars=['승차총승객수', '하차총승객수'])
df_total_sum_melted['구분'] = df_total_sum_melted['구분'].replace({'승차총승객수': '승차', '하차총승객수': '하차'})

plt.figure(figsize=(18, 8))
sns.barplot(x='호선명', y='value', hue='구분', data=df_total_sum_melted.sort_values(by='value', ascending=False), palette='viridis')
plt.title('호선별 승/하차총승객수 합계 비교', fontsize=18)
plt.xlabel('호선명', fontsize=12)
plt.ylabel('총 승객수', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='구분')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 4. 호선별 승/하차총승객수 평균 비교 막대 그래프 (두 개의 막대)
# 데이터를 long format으로 변환
df_grouped_avg_melted = df_grouped_avg.melt(id_vars='호선명', var_name='구분', value_vars=['승차총승객수', '하차총승객수'])
df_grouped_avg_melted['구분'] = df_grouped_avg_melted['구분'].replace({'승차총승객수': '승차 평균', '하차총승객수': '하차 평균'})

plt.figure(figsize=(18, 8))
sns.barplot(x='호선명', y='value', hue='구분', data=df_grouped_avg_melted.sort_values(by='value', ascending=False), palette='magma')
plt.title('호선별 일평균 승/하차총승객수 비교', fontsize=18)
plt.xlabel('호선명', fontsize=12)
plt.ylabel('일평균 승객수', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='구분')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
#혼잡도 임시 시각화

# 1. 호선별 총 승객 수 계산 (혼잡도 proxy)
df_total_sum['총승객수'] = df_total_sum['승차총승객수'] + df_total_sum['하차총승객수']

# 2. '총승객수' 기준으로 내림차순 정렬
df_congestion = df_total_sum.sort_values(by='총승객수', ascending=False).reset_index(drop=True)

print("\n--- 호선별 총 승객수 (혼잡도 지표) ---")
print(df_congestion[['호선명', '총승객수']])

# 3. 시각화: 막대 그래프
plt.figure(figsize=(14, 8)) # 그래프 크기 설정
sns.barplot(x='총승객수', y='호선명', data=df_congestion, palette='viridis') # Seaborn으로 막대 그래프 그리기
plt.title('호선별 총 승객 수 (혼잡도 지표)', fontsize=18)
plt.xlabel('총 승객 수', fontsize=14)
plt.ylabel('호선명', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7) # x축에만 그리드 표시
plt.tight_layout() # 그래프 요소들이 겹치지 않게 자동으로 레이아웃 조정
plt.show()

# 만약 승차 승객수와 하차 승객수를 같이 보고 싶다면 이런 방법도 있어!
plt.figure(figsize=(14, 8))
# 넓은 형식으로 데이터프레임 재구성 (seaborn stack bar chart를 위해)
df_melted = df_total_sum.melt(id_vars='호선명', value_vars=['승차총승객수', '하차총승객수'],
                             var_name='유형', value_name='승객수')
# 총 승객수를 기준으로 정렬된 순서를 유지하기 위해 호선명 순서를 지정
order_lines = df_congestion['호선명'].tolist()
sns.barplot(x='승객수', y='호선명', hue='유형', data=df_melted, order=order_lines, palette='magma')
plt.title('호선별 승차 및 하차 총 승객 수', fontsize=18)
plt.xlabel('승객 수', fontsize=14)
plt.ylabel('호선명', fontsize=14)
plt.legend(title='승객 유형')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()


plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 안 깨지게 하는 코드
from matplotlib import rc
rc('font', family='Malgun Gothic') # Windows 운영체제용
# plt.rcParams['font.family'] = 'AppleGothic' # Mac 운영체제용
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 데이터셋 파일 경로
excel_file_path = 'C:\\Users\\rkffl\\SubwayPulse\\1.csv'

# 파일을 불러올 때 인코딩 및 파일 존재 여부 오류 처리
try:
    df = pd.read_csv(excel_file_path, encoding='cp949')
    print("CSV 파일을 cp949 인코딩으로 성공적으로 읽었습니다.")
except UnicodeDecodeError:
    print("cp949 인코딩으로 파일을 읽는 데 실패했습니다. utf-8로 다시 시도합니다.")
    try:
        df = pd.read_csv(excel_file_path, encoding='utf-8')
        print("CSV 파일을 utf-8 인코딩으로 성공적으로 읽었습니다.")
    except UnicodeDecodeError:
        print("utf-8 인코딩으로도 파일을 읽는 데 실패했습니다. 다른 인코딩을 시도해 보세요.")
        exit() # 파일 인코딩 문제로 읽지 못하면 프로그램 종료
except FileNotFoundError:
    print(f"오류: '{excel_file_path}' 파일을 찾을 수 없습니다. 파일 경로와 이름을 다시 확인해주세요.")
    exit() # 파일을 찾지 못하면 프로그램 종료

# pandas 설정 변경: 모든 행과 모든 열을 출력하도록 설정
pd.set_option('display.max_rows', None)     # 최대 행 수 제한 해제 (모든 행 출력)
pd.set_option('display.max_columns', None)  # 최대 열 수 제한 해제 (모든 열 출력)
pd.set_option('display.width', 1000)        # 출력 너비 설정 (콘솔 출력 시 줄 바꿈 방지)


print("\n[단계 1/4] 데이터프레임 내용 (상위 5줄):")
print(df.head())

print("\n[단계 2/4] 현재 데이터프레임의 전체 컬럼 목록:")
print(df.columns.tolist()) # 컬럼 목록을 리스트 형태로 출력하여 보기 쉽게 함

# =========================================================================
# ****** ⭐⭐ 중요! 여기를 꼭 확인하고 네 CSV 파일의 실제 컬럼명으로 수정해야 해! ⭐⭐ ******
# 예를 들어, df.columns 결과에 ['노선명', '승차인원', '하차인원']이 나왔다면, 아래와 같이 변경
# LINE_COL = '노선명'
# ON_RIDE_COL = '승차인원'
# OFF_RIDE_COL = '하차인원'
# *************************************************************************
LINE_COL = '호선명'      # 실제 '호선명' 컬럼의 이름으로 변경 (예: '노선명')
ON_RIDE_COL = '승차총승객수' # 실제 '승차총승객수' 컬럼의 이름으로 변경 (예: '승차인원')
OFF_RIDE_COL = '하차총승객수' # 실제 '하차총승객수' 컬럼의 이름으로 변경 (예: '하차인원')
# =========================================================================

# --- 데이터 가공 및 기초 통계 ---
print(f"\n[단계 3/4] '{LINE_COL}', '{ON_RIDE_COL}', '{OFF_RIDE_COL}' 컬럼을 사용하여 데이터 처리 시작...")

# 'LINE_COL'을 기준으로 그룹화하고, 각 그룹의 'ON_RIDE_COL' 합계 구하기
df_grouped_sum_on = df.groupby(LINE_COL)[ON_RIDE_COL].sum().reset_index()
# 'LINE_COL'을 기준으로 그룹화하고, 각 그룹의 'OFF_RIDE_COL' 합계 구하기
df_grouped_sum_off = df.groupby(LINE_COL)[OFF_RIDE_COL].sum().reset_index()

# 호선별 승/하차총승객수 합계를 하나의 DataFrame으로 합치기
df_total_sum = pd.merge(df_grouped_sum_on, df_grouped_sum_off, on=LINE_COL)

# 'LINE_COL'을 기준으로 그룹화하고, 여러 컬럼의 평균 구하기
df_grouped_avg = df.groupby(LINE_COL)[[ON_RIDE_COL, OFF_RIDE_COL]].mean().reset_index()


# 승차 합계 출력
print(f"\n--- 호선별 {ON_RIDE_COL} 합계 ---")
print(df_grouped_sum_on.sort_values(by=ON_RIDE_COL, ascending=False))

# 하차 합계 출력
print(f"\n--- 호선별 {OFF_RIDE_COL} 합계 ---")
print(df_grouped_sum_off.sort_values(by=OFF_RIDE_COL, ascending=False))

# 승하차 평균 출력
print(f"\n--- 호선별 {ON_RIDE_COL}, {OFF_RIDE_COL} 평균 ---")
print(df_grouped_avg.sort_values(by=ON_RIDE_COL, ascending=False))


print("\n[단계 4/4] 데이터 시각화 시작...")

# --- 시각화 1: 호선별 승차총승객수 합계 막대 그래프 ---
plt.figure(figsize=(15, 7)) # 그래프 크기 설정
sns.barplot(x=LINE_COL, y=ON_RIDE_COL, data=df_grouped_sum_on.sort_values(by=ON_RIDE_COL, ascending=False), palette='Blues_d')
plt.title(f'호선별 {ON_RIDE_COL} 합계', fontsize=18)
plt.xlabel(LINE_COL, fontsize=12)
plt.ylabel(ON_RIDE_COL, fontsize=12)
plt.xticks(rotation=45, ha='right') # 호선명 라벨이 겹치지 않게 회전
plt.grid(axis='y', linestyle='--', alpha=0.7) # 그리드 라인 추가
plt.tight_layout() # 그래프 요소들이 잘 보이도록 자동 조정
plt.show()

# --- 시각화 2: 호선별 하차총승객수 합계 막대 그래프 ---
plt.figure(figsize=(15, 7))
sns.barplot(x=LINE_COL, y=OFF_RIDE_COL, data=df_grouped_sum_off.sort_values(by=OFF_RIDE_COL, ascending=False), palette='Reds_d')
plt.title(f'호선별 {OFF_RIDE_COL} 합계', fontsize=18)
plt.xlabel(LINE_COL, fontsize=12)
plt.ylabel(OFF_RIDE_COL, fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- 시각화 3: 호선별 승/하차총승객수 합계 비교 막대 그래프 (두 개의 막대) ---
df_total_sum_melted = df_total_sum.melt(id_vars=LINE_COL, var_name='구분', value_vars=[ON_RIDE_COL, OFF_RIDE_COL])
df_total_sum_melted['구분'] = df_total_sum_melted['구분'].replace({ON_RIDE_COL: '승차', OFF_RIDE_COL: '하차'})

plt.figure(figsize=(18, 8))
# 전체 합계를 기준으로 정렬된 호선명을 사용하여 그래프 순서 유지
order_sum_lines = df_total_sum_melted.groupby(LINE_COL)['value'].sum().sort_values(ascending=False).index
sns.barplot(x=LINE_COL, y='value', hue='구분', data=df_total_sum_melted, order=order_sum_lines, palette='viridis')
plt.title('호선별 승/하차총승객수 합계 비교', fontsize=18)
plt.xlabel(LINE_COL, fontsize=12)
plt.ylabel('총 승객수', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='구분')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- 시각화 4: 호선별 승/하차총승객수 평균 비교 막대 그래프 (두 개의 막대) ---
df_grouped_avg_melted = df_grouped_avg.melt(id_vars=LINE_COL, var_name='구분', value_vars=[ON_RIDE_COL, OFF_RIDE_COL])
df_grouped_avg_melted['구분'] = df_grouped_avg_melted['구분'].replace({ON_RIDE_COL: '승차 평균', OFF_RIDE_COL: '하차 평균'})

plt.figure(figsize=(18, 8))
# 평균 합계를 기준으로 정렬된 호선명을 사용하여 그래프 순서 유지
order_avg_lines = df_grouped_avg_melted.groupby(LINE_COL)['value'].sum().sort_values(ascending=False).index
sns.barplot(x=LINE_COL, y='value', hue='구분', data=df_grouped_avg_melted, order=order_avg_lines, palette='magma')
plt.title('호선별 일평균 승/하차총승객수 비교', fontsize=18)
plt.xlabel(LINE_COL, fontsize=12)
plt.ylabel('일평균 승객수', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='구분')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- 시각화 5: 호선별 총 승객 수 (혼잡도 지표) 막대 그래프 ---
df_total_sum['총승객수'] = df_total_sum[ON_RIDE_COL] + df_total_sum[OFF_RIDE_COL]
df_congestion_base = df_total_sum.sort_values(by='총승객수', ascending=False).reset_index(drop=True)

print("\n--- 호선별 총 승객수 (혼잡도 지표) ---")
print(df_congestion_base[[LINE_COL, '총승객수']])

plt.figure(figsize=(14, 8))
sns.barplot(x='총승객수', y=LINE_COL, data=df_congestion_base, palette='viridis')
plt.title('호선별 총 승객 수 (혼잡도 지표)', fontsize=18)
plt.xlabel('총 승객 수', fontsize=14)
plt.ylabel(LINE_COL, fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- 시각화 6: 호선별 승차 및 하차 총 승객 수 함께 비교 ---
plt.figure(figsize=(14, 8))
df_melted_total = df_total_sum.melt(id_vars=LINE_COL, value_vars=[ON_RIDE_COL, OFF_RIDE_COL],
                             var_name='유형', value_name='승객수')
order_lines_total = df_congestion_base[LINE_COL].tolist() # 총 승객수 기준 정렬 순서 유지
sns.barplot(x='승객수', y=LINE_COL, hue='유형', data=df_melted_total, order=order_lines_total, palette='magma')
plt.title('호선별 승차 및 하차 총 승객 수', fontsize=18)
plt.xlabel('승객 수', fontsize=14)
plt.ylabel(LINE_COL, fontsize=14)
plt.legend(title='승객 유형')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# =========================================================================
# ⭐⭐⭐ 혼잡도 4단계 분류 시각화 ⭐⭐⭐
# 실제 혼잡도 공식 데이터가 아닌, '총 승객수'를 기준으로 가상 혼잡도를 계산하여 분류합니다.
# =========================================================================

print("\n--- 호선별 총 승객수를 기반으로 한 가상 혼잡도 (4단계) ---")

# '총승객수'를 퍼센티지 스케일로 변환하기 위한 가상 정원 설정
# 가장 혼잡한 호선이 대략 200%의 혼잡도를 가지도록 임의의 스케일을 조정 (이 숫자는 필요에 따라 변경 가능)
max_total_passengers = df_congestion_base['총승객수'].max()
HYPOTHETICAL_CAPACITY_FACTOR = max_total_passengers / 200 # 최고 총 승객수를 200%로 가정했을 때의 1%에 해당하는 값

# '가상_혼잡도(%)' 컬럼 생성
df_congestion_base['가상_혼잡도(%)'] = (df_congestion_base['총승객수'] / HYPOTHETICAL_CAPACITY_FACTOR)

# 혼잡도 4단계 분류 함수 정의
def classify_congestion(percentage):
    if percentage >= 170:
        return '4단계 (심각)'
    elif percentage >= 150: # 150% 이상 ~ 169%
        return '3단계 (매우 혼잡)'
    elif percentage >= 130: # 130% 이상 ~ 149%
        return '2단계 (혼잡)'
    else: # 129% 이하
        return '1단계 (주의)'

# 새로운 컬럼에 혼잡도 단계 적용
df_congestion_base['혼잡도_단계'] = df_congestion_base['가상_혼잡도(%)'].apply(classify_congestion)

# 혼잡도_단계에 따른 정렬 순서 정의
congestion_order = ['1단계 (주의)', '2단계 (혼잡)', '3단계 (매우 혼잡)', '4단계 (심각)']
df_congestion_base['혼잡도_단계'] = pd.Categorical(df_congestion_base['혼잡도_단계'], categories=congestion_order, ordered=True)

# 단계별 색상 지정 (seaborn palette는 자동으로 색상 할당하지만, 명확하게 지정하고 싶을 때 사용)
palette_map = {
    '1단계 (주의)': 'lightgreen',
    '2단계 (혼잡)': 'gold',        # 노란색 계열
    '3단계 (매우 혼잡)': 'darkorange', # 주황색 계열
    '4단계 (심각)': 'red'
}

# --- 시각화 7: 호선별 가상 혼잡도 (4단계 분류) 막대 그래프 ---
plt.figure(figsize=(14, 8))
# 가상 혼잡도가 높은 순서대로 정렬
sns.barplot(x='가상_혼잡도(%)', y=LINE_COL, hue='혼잡도_단계',
            data=df_congestion_base.sort_values(by='가상_혼잡도(%)', ascending=False),
            palette=palette_map, dodge=False) # dodge=False로 hue 카테고리별로 막대가 분리되지 않고 하나의 막대에 색상으로 표현되게 함
plt.title('호선별 가상 혼잡도 (4단계 분류)', fontsize=18)
plt.xlabel('가상 혼잡도 (%)', fontsize=14)
plt.ylabel(LINE_COL, fontsize=14)
plt.xlim(0, max(df_congestion_base['가상_혼잡도(%)']) * 1.1) # x축 범위를 최대값보다 약간 크게 설정
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- 시각화 8: 혼잡도 단계별 호선 수 카운트 그래프 ---
plt.figure(figsize=(10, 6))
# 혼잡도_단계 순서에 따라 정렬하고, 단계별 색상 적용
sns.countplot(x='혼잡도_단계', data=df_congestion_base, order=congestion_order, palette=palette_map)
plt.title('혼잡도 단계별 호선 수', fontsize=18)
plt.xlabel('혼잡도 단계', fontsize=14)
plt.ylabel('호선 수', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n모든 시각화가 완료되었습니다!")
```

```python
# 혼잡도 2 _다른사람이 만든거

# 데이터 불러오기
df = pd.read_csv('2.csv', encoding='cp949')  # 실제 파일명과 경로 입력

# 시간대 컬럼 리스트 (받은 시간대 데이터 예시 참고)
time_columns = ['5시30분', '6시00분', '6시30분', '7시00분', '7시30분', '8시00분', '8시30분',
                '9시00분', '9시30분', '10시00분', '10시30분', '11시00분', '11시30분', '12시00분',
                '12시30분', '13시00분', '13시30분', '14시00분', '14시30분', '15시00분', '15시30분',
                '16시00분', '16시30분', '17시00분', '17시30분', '18시00분', '18시30분', '19시00분',
                '19시30분', '20시00분', '20시30분', '21시00분', '21시30분', '22시00분', '22시30분',
                '23시00분', '23시30분', '00시00분', '00시30분']

# 데이터프레임을 긴 형태로 변환
df_long = df.melt(id_vars=['요일구분', '호선', '역번호', '출발역', '상하구분'],
                  value_vars=time_columns,
                  var_name='시간대',
                  value_name='혼잡도')

# 혼잡도 데이터가 퍼센트라면 숫자 타입으로 변환
df_long['혼잡도'] = pd.to_numeric(df_long['혼잡도'], errors='coerce')

# 예시 1: 평일과 주말 혼잡도 시간대별 평균 비교 라인 그래프
plt.figure(figsize=(14,7))
sns.lineplot(data=df_long, x='시간대', y='혼잡도', hue='요일구분', estimator='mean', ci=None)
plt.xticks(rotation=45)
plt.title('요일구분별 시간대별 평균 혼잡도')
plt.ylabel('혼잡도 (%)')
plt.xlabel('시간대')
plt.grid(True)
plt.tight_layout()
plt.show()

# 예시 2: 특정 호선 평일 시간대별 혼잡도 히트맵
line_to_plot = '2호선'  # 원하는 호선으로 변경 가능
pivot_table = df_long[(df_long['호선'] == line_to_plot) & (df_long['요일구분'] == '평일')].pivot_table(
    index='출발역', columns='시간대', values='혼잡도', aggfunc='mean')

plt.figure(figsize=(16,10))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar_kws={'label': '혼잡도 (%)'})
plt.title(f'{line_to_plot} 평일 출발역별 시간대별 평균 혼잡도 히트맵')
plt.ylabel('출발역')
plt.xlabel('시간대')
plt.tight_layout()
plt.show()

# 예시 3: 특정 시간대 집중, 호선별 혼잡도 막대 그래프
target_time = '10시30분'  # 집중할 시간대 지정
df_time = df_long[df_long['시간대'] == target_time].groupby('호선')['혼잡도'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=df_time, x='호선', y='혼잡도', palette='viridis')
plt.title(f'{target_time} 기준 호선별 평균 혼잡도')
plt.ylabel('혼잡도 (%)')
plt.xlabel('호선')
plt.tight_layout() 
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# MNIST 데이터셋을 읽고 신경망에 입력할 형태로 변환
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 28, 28, 1)
x_test = x_test.reshape(10000, 28, 28, 1)
x_train = x_train.astype(np.float32) / 255.0
x_test = x_test.astype(np.float32) / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# 신경망 모델 설계
cnn = Sequential()
cnn.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
cnn.add(Conv2D(64, (3, 3), activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Flatten())
cnn.add(Dense(128, activation='relu'))
cnn.add(Dropout(0.5))
cnn.add(Dense(10, activation='softmax'))

# ------------------------------------------------------------------
# 1. 모델 컴파일 (오류 해결)
# ------------------------------------------------------------------
cnn.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# ------------------------------------------------------------------
# 2. 모델 학습 (hist 변수 정의 및 그래프 출력을 위해 필수)
# ------------------------------------------------------------------
print("모델 학습 시작...")
hist = cnn.fit(x_train, y_train, batch_size=128, epochs=10, validation_data=(x_test, y_test), verbose=1)
print("모델 학습 완료.")
# ------------------------------------------------------------------


# 신경망 모델 정확률 평가
res = cnn.evaluate(x_test, y_test, verbose=0)
print("\n테스트 데이터셋 정확률:", res[1] * 100)

# 정확률 그래프
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='best')
plt.grid()

# 손실 함수 그래프
plt.subplot(1, 2, 2)
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='best')
plt.grid()

plt.tight_layout()
plt.show()
```
